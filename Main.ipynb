```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("Bitcoin History.csv")
```

```python
total_length = len(data)
print("Total Length of the Dataset:", total_length)
```

```python
data['Date'] = pd.to_datetime(data['Date'], format='%b %d, %Y')
data['Price'] = data['Price'].str.replace(',', '').astype(float)

data['Year'] = data['Date'].dt.year
average_price_per_year = data.groupby('Year')['Price'].mean()
plt.figure(figsize=(10, 6))
average_price_per_year.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Average Price of Bitcoin')
plt.xlabel('Year')
plt.ylabel('Average Price')
plt.xticks(rotation=45)
for i, value in enumerate(average_price_per_year):
    plt.text(i, value, f'{value:.2f}', ha='center', va='bottom', fontsize=8)

plt.grid(True)
plt.tight_layout()
plt.show()
```
